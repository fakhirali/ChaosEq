```python
import random
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
def eq(rate, x):
    next_x = rate*x[-1]*(1-x[-1])
    x.append(next_x)
```

```python
starting_num = random.random()
starting_num_arr  = [starting_num]
print(starting_num_arr)
rate = random.random()*4
print(rate)
for i in range(499):
    eq(rate,starting_num_arr)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()
y = starting_num_arr
x = np.arange(0, 500)

line, = ax.plot(x, y)


def animate(i):
    line.set_ydata(y[:i])# update the data
    line.set_xdata(x[:i])
    xmin, xmax = ax.get_xlim()
    ymin, ymax = ax.get_ylim()
    if x[i] >= xmax:
        ax.set_xlim(xmax/2, 2*xmax)
        ax.figure.canvas.draw
    return line,


# Init only required for blitting to give a clean slate.
def init():
    ax.set_ylim(0, 1)
    ax.set_xlim(0, 10)
    line.set_ydata(np.ma.array(x, mask=True))
    return line,

ani = animation.FuncAnimation(fig, animate, np.arange(1,500), init_func=init,
                              interval=200, blit=True)
plt.show()
```
